```python
rows = []

for test in TEST_SET:
    prompt = base_generated_prompt.render(
        query=test["query"],
        chat_history=test["chat_history"],
    )
    response = openai_generate(prompt)
    rows.append(
        {
            "query": test["query"],
            "greet": test["greet"],
            "chat_history": test["chat_history"].strip(),
            "response": response,
        }
    )

report_df = pd.DataFrame(rows)
report_path = save_base_prompting_report(
    report_df=report_df,
    output_path=Path("../reports/base_prompting_report.md"),
)

display(HTML(f"<p><strong>Report saved to:</strong> {escape(str(report_path))}</p>"))
report_df
```
